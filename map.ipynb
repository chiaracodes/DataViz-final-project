```python
countries = alt.topo_feature(data.world_110m.url, 'countries')

slider = alt.binding_range(name='Year',
    step=1,
    min=board_members['Year'].min(), 
    max=board_members['Year'].max()
)

select_date = alt.selection_single(
    name="slider", 
    fields=['Year'],
    bind=slider, 
)


alt.Chart(board_members).mark_geoshape()\
    .encode(color='Value:Q',
            tooltip=['Country:N', 'Value:Q']
           )\
    .add_selection(select_date)\
    .transform_filter(select_date)\
    .transform_lookup(
        lookup='id',
        from_=alt.LookupData(countries, key='id',
                             fields=["type", "properties", "geometry"])
    )\
    .project('mercator',
            scale=300,
         center=[20,50],
         clipExtent= [[0, 0], [400, 300]])\
    .properties(
        width=500,
        height=300,
        title='Europe Female Share of Seats'
    )
```
